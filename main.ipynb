#### What's the mean salary by department?
Hint : There are 2 ways to do it - one way is with groupby and the other way is with pivot tables. Do it both ways. 


```python
# Method 1 Pivot Table
employee_salary = employee

employee_salary.pivot_table(index=['Department'])
```

```python
#Method 2 groupby
employee_salary.groupby(['Department']).agg({'Salary':np.mean}).reset_index().head()
```

#### We want now to group the data by more than one column. Use Pandas pivot table function to select the mean salary by department and title.

```python
#Method 1 group by
employee_salary.groupby(['Department', 'Title']).agg({'Salary':np.mean}).reset_index().head()
```

```python
# Method 2 pivot table
employee_salary.pivot_table(index=['Department', 'Title'], values=['Salary'])
```
